```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.callbacks import ModelCheckpoint
from datetime import datetime 

num_epochs = 100
num_batch_sizes = [8, 12, 16, 32]

model.compile(loss='categorical_crossentropy',metrics=['accuracy', 'mean_squared_error'],optimizer='Nadam')

# Dictionary to store training durations for each batch size
training_durations = {}

# Assuming X_train, y_train, X_val, and y_val are properly preprocessed and have the right shapes

# Define your model architecture
# model = Sequential([
#     # Replace ... with your actual input shape
#     Flatten(input_shape=(...)),
#     # Add more layers as needed
#     Dense(128, activation='relu'),
#     Dense(64, activation='relu'),
#     Dense(num_classes, activation='softmax')  # Replace num_classes with the number of classes in your classification task
# ])


# Iterate over each batch size
for batch_size in num_batch_sizes:
    checkpointer = ModelCheckpoint(filepath=f'saved_models/audio_classification_{batch_size}.hdf5', 
                                   verbose=1, save_best_only=True)
    start = datetime.now()

    # Fit the model for the current batch size
    model.fit(X_train, y_train, batch_size=batch_size, epochs=num_epochs, 
              validation_data=(X, y), callbacks=[checkpointer], verbose=1)
    
    duration = datetime.now() - start
    training_durations[batch_size] = duration.total_seconds()

    print(f"Training completed for batch size {batch_size} in time: {duration}")

# Print training durations
for batch_size, duration in training_durations.items():
    print(f"Batch size: {batch_size}, Training duration (seconds): {duration}")
```

```python
# Dictionary to store accuracies for each batch size
accuracies = {}

# Iterate over each batch size
for batch_size in num_batch_sizes:
    # Load the best model for the current batch size
    model.load_weights(f'saved_models/audio_classification_{batch_size}.hdf5')
    
    # Evaluate the model on test data
    accuracy = model.evaluate(X_test, y_test, batch_size=batch_size, verbose=0)
    
    # Store the accuracy for the current batch size
    accuracies[batch_size] = accuracy

# Print accuracies
for batch_size, accuracy in accuracies.items():
    print(f"Batch size: {batch_size}, Accuracy: {accuracy[1]*100}")
```
